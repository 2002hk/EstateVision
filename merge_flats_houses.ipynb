```python
import pandas as pd
import numpy as np
```

```python
flats = pd.read_csv('/content/flats_cleaned2 (1).csv')
houses = pd.read_csv('/content/house_cleaned.csv')
```

```python
df = pd.concat([flats,houses],ignore_index=True)
```
