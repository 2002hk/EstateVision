```python
df=pd.read_csv('/content/independent-house - independent-house.csv')
df.head()
```

```python
df=df.drop('link',axis=1)
```

```python
df.rename(columns={'rate':'price_per_sqft'},inplace=True)
```

```python
df=df[df['price']!='Price on Request']
```

```python
#converting price to crores
def price_to_cr(x):
    # Check if the price is in 'Lac' or 'Crore' and convert accordingly
    try:
        if x[1] == 'Lac':
            # Convert Lac to base (e.g., in lakhs)
            return round(float(x[0])/100, 2)  # Keep as is, since 1 Lac = 1,00,000
        elif x[1] == 'Crore':
            # Convert Crore to base (e.g., multiply by 100, as 1 Crore = 100 Lac)
            return round(float(x[0]), 2)
    except (ValueError, IndexError):
        # Handle cases where conversion fails or unexpected format
        return None
```

```python
df[df['price'].isna()].shape
```

```python
df['price'].dropna(inplace=True)
```

```python
df['price'] = df['price'].astype(str)
```

```python
df['price']=df['price'].str.split(' ').apply(price_to_cr)
df['price'].head(15)
```

```python
#fetching price per square feet value
df['price_per_sqft'] = (
    df['price_per_sqft']
    .str.split('/').str.get(0)   # Split on '/' and get the first part
    .str.replace('₹', '')        # Remove the '₹' symbol
    .str.replace(',', '')        # Remove commas from the numeric strings
    .str.strip()                 # Strip any leading/trailing whitespace
)

# Convert to float, coercing errors (non-convertible strings will become NaN)
df['price_per_sqft'] = pd.to_numeric(df['price_per_sqft'], errors='coerce')

# Display rows with NaN (optional, to inspect invalid entries)
```
